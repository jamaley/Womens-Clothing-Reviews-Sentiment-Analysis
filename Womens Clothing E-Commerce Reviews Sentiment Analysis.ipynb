```python
import pandas as pd
import os
from textblob import TextBlob
from textblob import Word
```

```python
data = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
data.head()
```

```python
data['Review Text'] = data['Review Text'].astype(str)
```

```python
data['polarity'] = data['Review Text'].map(lambda text: TextBlob(text).sentiment.polarity)
```

```python
data.dtypes
```

```python
data.head()
```

```python
data['review_len'] = data['Review Text'].astype(str).apply(len)
data['word_count'] = data['Review Text'].apply(lambda x: len(str(x).split()))
```

```python
data.head(5)
```

## Then randomly select 5 reviews with the most positive sentiment polarity score (one):

```python
print('5 random reviews with the highest positive sentiment polarity: \n')
positive = data.loc[data.polarity == 1, ['Review Text']].sample(5).values
for i,p in enumerate(positive):
    print(i,"-",p[0],"\n")
```

```python
print('5 random reviews with the highest positive sentiment polarity: \n')
positive = data.loc[data.polarity < 0, ['Review Text']].sample(5).values
for i,p in enumerate(positive):
    print(i,"-",p[0],"\n")
```
